```python
df = pd.read_csv("./mean.txt")
plt.figure()
sns.scatterplot(df, x="num",y="mean")
sns.scatterplot(df, x="num",y="abs_mean")
sns.scatterplot(df, x="num",y="sq_mean")

p=np.polyfit(df["abs_mean"], df["num"], 2)
xx=np.linspace(min(df["abs_mean"]), max(df["abs_mean"]), 100)
yy=np.polyval(p, xx)
plt.plot(yy, xx)

p=np.polyfit(df["sq_mean"], df["num"], 2)
xx=np.linspace(min(df["sq_mean"]), max(df["sq_mean"]), 100)
yy=np.polyval(p, xx)
plt.plot(yy, xx)
```

```python
plt.rcParams['text.usetex'] = True

with open("./limit.txt") as f:
    values = [int(round(float(i)/100)) for i in f.readline().split(', ')[:-1]]
    x = list(set(values))
    y = [values.count(i)/len(values) for i in x]
plt.bar(x, y)
plt.xlabel(r'value$, \cdot 10^{-2}$', fontsize=16, color='k')
plt.ylabel(r'probability')
```
