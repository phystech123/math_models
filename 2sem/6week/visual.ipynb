```python
plt.figure()
for j in range(900, 1000):
    r = np.linspace(0, 4, 100)
    a = []
    b = []
    for i in r:
        os.system(f'./build/main {i} 0.9')
        dfX = pd.read_csv("data/1.csv", sep=",")
        b.append(dfX.iloc[j, 1])
        a.append(i)
    sns.scatterplot(x=a, y=b, alpha=0.5, s=1)
plt.savefig("pic4.svg")
```
