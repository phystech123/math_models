```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dfX = pd.read_csv("dataX_eyler.csv", sep=";")
dfX
```

```python
plt.figure()
sns.scatterplot(data=dfX, x='t', y=' x')
xx = dfX['t'].to_numpy()
yy = np.sin(xx*10+np.pi/2)
sns.scatterplot(x = xx, y=yy)
```

```python
dfV = pd.read_csv("dataV_eyler.csv", sep=";")
dfV
```

```python
plt.figure()
sns.scatterplot(data=dfV, x='t', y=' x')
xx = dfV['t'].to_numpy()
yy = np.sin(xx*10)*(-10)
sns.scatterplot(x = xx, y=yy)
```
