```python
dir = './'
NUM_NEW_CONF = 1
NUM_EX_CONF = len(list(Path("configs").iterdir()))

dim = 2

for i in range(NUM_EX_CONF, NUM_EX_CONF + NUM_NEW_CONF):
    config = {"dim":dim, "params":[float(x) for x in input().split()], "conditions":[float(x) for x in input().split()], "fineness":float(input()), "time_limit":float(input()), "output_files":[]}
    for j in range(config["dim"]):
        config["output_files"].append(f"{dir}outputs/{j}.csv")
    with open(f"{dir}configs/config{i}.json", 'w') as file:
        json.dump(config, file)
```
